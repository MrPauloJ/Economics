Considere uma economia de dotação pequena, aberta e de dois períodos com bens de
consumo duráveis. A característica distintiva dessa economia é que as compras de bens de consumo
duráveis no período 1, C1, continuam a fornecer utilidade no período 2. O fluxo de serviços que
as famílias recebem do estoque de duráveis no período 2 depende de novas compras de duráveis no
período 2, C2, e da estoque não-depreciado de duráveis adquiridos no período 1. Assuma que os
bens de consumo duráveis depreciam à taxa δ ∈ [0, 1]. As preferências são descritas pela seguinte
função utilidade:

U(C1, C2) = ln(C1) + ln(C2 + (1 − δ)C1).

Continue a assumir que a economia experimenta choques de produto positivos e persistentes,
ou seja, Q1 aumenta em 1 e Q2 aumenta em ρ ∈ (0, 1). Mas agora não imponha que δ = 1.
Encontre os pares (δ, ρ) tais que a resposta da balança comercial no período 1 seja anticíclica,
ou seja, negativa, e as compras de consumo, C1 e C2, sejam positivas. Forneça intuição para

```python
# Definindo intervalo
rho=np.arange(0.00001,0.9999,0.01)
delta=np.arange(0,1,0.01)
```

```python
# Criando coordenadas
[rho,delta]=np.meshgrid(rho,delta)
```

```python
# Realizando cálculo e inserindo em uma matriz lógica
i = ([(2*delta-rho-0.9)/(0.2-2*delta) < 0])
```

```python
# Plotando gráfico
px.scatter(x=delta[i],y=rho[i], labels={"x":"Delta", "y":"Rho"})
```
