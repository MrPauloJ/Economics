Questão 1)

 Uma companhia erea tem o seguinte programa de milhagem: As passagens  ́
tem abatimento de 25% ao completar 10.000 milhas de voo em um ano e tem abatimento de 50% quando se completam 25.000 milhas de voo por ano.

Esboce graficamente a restrição orçamentária e mostre as equações correspondentes de um viajante desta companhia, ao planejar seus voos (y) por um ano. Assuma que (x) representa todos os outros gastos do viajante.

```python
#Quantidade de viagens
y = np.arange(0,50000,1)

#Preço de outros bens e de viagens
px=py=1

#Renda
m=50000

#Restrição orçamentaria de viagens e outros bens
def func(y):
    global px,py,m
    if (y<10000) :
        return (m/px)-((py*y)/px)
    if ((y>=10000) & (y<25000)) :
        return ((m-(0.25*10000))/px)-(((py*0.75*y)/px))
    if (y>=25000) :
        return ((m-(0.35*25000))/px)-(((py*0.5*y)/px))
    
#Valores de y na R.O
x = []
for l in y:
    x.append(func(l))

#Plotando gráfico
plt.plot(y,x)
plt.title("Restrição orçamentária")
plt.ylabel("Outros bens(x)")
plt.ylim(bottom=0)
plt.xlabel("Viagens(y)")
plt.show()
```

Questão 2)

Assuma um consumidor com função utilidade U = min{xa, xb} e preços Pa e
Pb.

A utilidade marginal de um dos bens e sempre igual a zero? Justifique a sua resposta assumindo quaisquer quantidades para xa e xb. 

De uma resposta completa através de gráficos.

```python
#Quantidades da função utilidade
xa=np.arange(0,100,1)
xb=np.arange(0,100,1)

#Criando os vetores no plano
XA,XB=np.meshgrid(xa,xb)

#Função utilidade
def utilidade1(parametros):
    x_a, x_b = parametros
    return np.minimum(x_a,x_b)

#Plotando gráfico
k=np.array(utilidade1([np.ravel(XA), np.ravel(XB)]))
z=k.reshape(XA.shape)
ax2 = plt.subplot()
curva = ax2.contour(XA, XB, z) #Curva de nível
ax2.set_ylabel("Bem B")
ax2.set_xlabel("Bem A")
ax2.set_ylim(bottom=0)
ax2.scatter(curva.levels[1],curva.levels[1], color="red", label="p=("+str(curva.levels[1])+","+str(curva.levels[1]))
ax2.scatter(20,15, color="blue", label="p=(20,15)")
plt.legend()
plt.show()

#Explicação
print("Dado que a utilidade é o valor minimo ao tratar-se de um bem complementar")
print("utilidade marginal de um bem será igual a zero pois ele consumirá uma quantidade proporcional entre os bens")
print("e uma unidade adicional em apenas um dos bem fora da proporção não dá ao consumidor uma utilidade maior")
```
