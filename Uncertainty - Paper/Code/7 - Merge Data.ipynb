```python
# Import libraries
import pandas as pd
```

```python
# Define the main path
path="..\\Data\\Processed\\Time Series\\"
```

```python
# Data to read
BVSP = pd.read_excel(str(path)+"BVSP.xlsx")
IIE = pd.read_excel(str(path)+"IIE-Br.xlsx")
selic = pd.read_excel(str(path)+"Selic_Meta.xlsx")
EPU = pd.read_excel(str(path)+"EPU_Br.xlsx")
GEPU = pd.read_excel(str(path)+"GEPU_PPP.xlsx")
PIM = pd.read_excel(str(path)+"PIM_PF_Desazonalizado.xlsx")
```

```python
# Aux variables
estados = {
    "Bahia":"BA",
    "Ceará":"CE",
    "Espírito Santo":"ES",
    "Goiás":"GO",
    "Minas Gerais":"MG",
    "Pará":"PA",
    "Paraná":"PR",
    "Pernambuco":"PE",
    "Rio de Janeiro":"RJ",
    "Rio Grande do Sul":"RS",
    "Santa Catarina":"SC",
    "São Paulo":"SP"
}

estado=[
    "Bahia",
    "Ceara",
    "Espirito Santo",
    "Goias",
    "Minas Gerais",
    "Para",
    "Parana",
    "Pernambuco",
    "Rio de Janeiro",
    "Rio Grande do Sul",
    "Santa Catarina",
    "Sao Paulo"
]

UF=["BA","CE","ES","GO","MG","PA","PR","PE","RJ","RS","SC","SP"]

pathState=[
    str(path)+"Bahia_IEF_IBC.xlsx",
    str(path)+"Ceara_IEF_IBC.xlsx",
    str(path)+"Espirito Santo_IEF_IBC.xlsx",
    str(path)+"Goias_IEF_IBC.xlsx",
    str(path)+"Minas Gerais_IEF_IBC.xlsx",
    str(path)+"Para_IEF_IBC.xlsx",
    str(path)+"Parana_IEF_IBC.xlsx",
    str(path)+"Pernambuco_IEF_IBC.xlsx",
    str(path)+"Rio de Janeiro_IEF_IBC.xlsx",
    str(path)+"Rio Grande do Sul_IEF_IBC.xlsx",
    str(path)+"Santa Catarina_IEF_IBC.xlsx",
    str(path)+"Sao Paulo_IEF_IBC.xlsx"
    ]
```

```python
# Put all data together
main = pd.DataFrame()
main["time"] = BVSP.index.values
main["BVSP"] = BVSP["BVSP"]
main["Selic"] = selic["Selic_Meta"]
main["EPU"] = EPU["EPU_Br"]
main["GEPU"] = GEPU["GEPU_PPP"]
main["IIE"] = IIE["IIE-Br"]
```

```python
# Include all states
temp_uf=main.copy()
for i in range(0,len(pathState),1):
    temp=pd.read_excel(pathState[i])
    temp_uf[str("IBCR."+UF[i])]=temp[str("IBCR."+UF[i]+".Desazonalizado")]
    temp_uf[str("IEF."+UF[i])]=temp[str("IEF."+UF[i]+".Desazonalizado")]
    temp_uf[str("PIM."+UF[i])]=PIM[str(estado[i])]
    temp_uf.to_excel(str(path)+"\\TUDO_general.xlsx")
```
