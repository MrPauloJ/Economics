Calculo do salário familia com base nas regras legais de 2019

```python
#importando banco
pnadc = pd.read_csv("/content/drive/MyDrive/Microsimulação/PNADC_2019_visita 5 (CSV).csv",sep=";",decimal=",",na_values=" ",dtype=float)
```

```python
#SALARIO FAMILIA (TRABALHO PRINCIPAL)
```

```python
#Selecionando menores de 14 anos que sejam 'Filho(a) do responsável e do cônjuge', ou, 'Filho(a) somente do responsável', ou, Enteado(a) - elegiveis
pnadc.loc[(pnadc["V2009"]<=14) & ((pnadc["V2005"]==4) | (pnadc["V2005"]==5) | (pnadc["V2005"]==6)),"V2005_14"]=1
```

```python
#Agregando variavel no banco de dados
pnadc=pnadc.join(pnadc.groupby("CHAVE.D")["V2005_14"].sum(),on="CHAVE.D",rsuffix="_bin")
pnadc.rename(columns={"V2005_14_bin":"FILHO14"},inplace=True)
```

```python
#Filtrando e obtendo salário familia - 2 faixa
pnadc.loc[((pnadc["V4012"]==1) | (pnadc["V4012"]==3)) & (pnadc["V4029"]==1) & \
          (((pnadc["V403312"] - (pnadc["FILHO14"] * 32.80) )>= 907.78) & \
           ((pnadc["V403312"] - (pnadc["FILHO14"] * 32.80))<=1364.43)),"SALFAM1"] \
          = pnadc["FILHO14"] * 32.80
```

```python
#Filtrando e obtendo salário familia - 1 faixa
pnadc.loc[((pnadc["V4012"]==1) | (pnadc["V4012"]==3)) & (pnadc["V4029"]==1)  & \
          ((pnadc["V403312"] - (pnadc["FILHO14"] * 46.54))<=907.77),"SALFAM1"]\
           = pnadc["FILHO14"] * 46.54
```

```python
#Ajustando salário familia apenas para responsáveis
pnadc.loc[(pnadc["V2005"]>=4) & (pnadc["V2005"]<=19),"SALFAM1"]=0
```

```python
#SALARIO FAMILIA (TRABALHO SECUNDARIO)
```

```python
#Filtrando e obtendo salário familia - 2 faixa
pnadc.loc[((pnadc["V4043"])==1 | (pnadc["V4043"]==3)) & (pnadc["V4048"]==1)  & \
          (((pnadc["V405012"] - (pnadc["FILHO14"] * 32.80))>=907.78) &
           ((pnadc["V405012"] - (pnadc["FILHO14"] * 32.80))<=1364.43)),"SALFAM2"]\
          = pnadc["FILHO14"] * 32.80
```

```python
#Filtrando e obtendo salário familia - 1 faixa
pnadc.loc[((pnadc["V4043"]==1) | (pnadc["V4043"]==3)) & (pnadc["V4048"]==1)  & \
          ((pnadc["V405012"] - (pnadc["FILHO14"] * 46.54))<=907.77),"SALFAM2"]\
          =pnadc["FILHO14"] * 46.54
```

```python
#Ajustando salário familia apenas para responsáveis
pnadc.loc[(pnadc["V2005"]>=4) & (pnadc["V2005"]<=19),"SALFAM2"]=0
```

```python
#SALARIO FAMILIA (APOSENTADOS)
```

```python
#Filtrando e obtendo salário familia - 2 faixa - Homens
pnadc.loc[(pnadc["V5004A2"]>0) & (pnadc["V2009"]>=65) & (pnadc["V2007"]== 1) & \
          (((pnadc["V5004A2"] - (pnadc["FILHO14"] * 32.80))>=907.78) & \
          ((pnadc["V5004A2"] - (pnadc["FILHO14"] * 32.80))<=1364.43)),"SALFAM3"] \
          =pnadc["FILHO14"] * 32.80
```

```python
#Filtrando e obtendo salário familia - 1 faixa - Homens
pnadc.loc[(pnadc["V5004A2"]>0) & (pnadc["V2009"]>=65) & (pnadc["V2007"]== 1) & \
          ((pnadc["V5004A2"] - (pnadc["FILHO14"] * 46.54))<=907.77),"SALFAM3"] \
          =pnadc["FILHO14"] * 46.54
```

```python
#Filtrando e obtendo salário familia - 2 faixa - Mulheres
pnadc.loc[(pnadc["V5004A2"]>0) & (pnadc["V2009"]>=60) & (pnadc["V2007"]== 2) & \
          (((pnadc["V5004A2"] - (pnadc["FILHO14"] * 32.80))>=907.78) & \
          ((pnadc["V5004A2"] - (pnadc["FILHO14"] * 32.80))<=1364.43)),"SALFAM3"] \
          =pnadc["FILHO14"] * 32.80
```

```python
#Filtrando e obtendo salário familia - 1 faixa - Mulheres
pnadc.loc[(pnadc["V5004A2"]>0) & (pnadc["V2009"]>=60) & (pnadc["V2007"]== 2) & \
          ((pnadc["V5004A2"] - (pnadc["FILHO14"] * 46.54))<=907.77),"SALFAM3"] \
          =pnadc["FILHO14"] * 46.54
```

```python
#Ajustando salário familia apenas para responsáveis
pnadc.loc[(pnadc["V2005"]>=4) & (pnadc["V2005"]<=19),"SALFAM3"]=0
```

```python
#SALARIO FAMILIA (TRABALHO PRINCIPAL + SECUNDARIO + APOSENTADOS)
```

```python
#Colocando 0 em 'missings'
pnadc.fillna(0,inplace=True)
```

```python
#Conhecendo salário familia total
pnadc["SALFAM"]=pnadc["SALFAM1"]+pnadc["SALFAM2"]+pnadc["SALFAM3"]
```

```python
#SEGURO DESEMPREGO
```

* OBS: A simulação é feita supondo que o benefício é recebido por pessoas desocupadas que têm rendas diversas entre o piso e o teto dos valores da tabela oficial do seguro desemprego.
* OBS: Para gerar um numero total de segurados próximo dos dados oficiais de 2019, complementa-se com uma seleção aleatória entre o restante das pessoas desocupadas que não passaram pelo primeiro filtro.
* OBS: O valor do benefício é determinado utilizando-se o valor médio oficial do benefício em 09/2019 (R$1.267,11) e do número de parcelas recebidas (4,78), último ano para o qual há dados disponíveis.

```python
#Criando variável de seguro desemprego de acordo com o valor médio e quantidade de parcelas
pnadc.loc[(pnadc["VD4002"]==2) & ((pnadc["RDIV"]>=0.8*998) & (pnadc["RDIV"]<=1735.29)), "SEGDEMP"] = (1267.11*4.78)/12
```

```python
#Organizando e obtendo amostra de 53% de desempregados
dados2=pnadc.groupby("CHAVE.I")["VD4002"].first()
dados2=pd.DataFrame(dados2)
dados2=dados2.loc[dados2["VD4002"]==2].sample(frac=0.53)
#Variavel temporaria
dados2["TEMP"]=1
```

```python
#Excluindo variavel do dataset original
pnadc.drop(columns=["VD4002"],inplace=True)
```

```python
#Criando novo dataset com a mesclagem da amostra e banco de dados original
dados3 = pnadc.join(dados2,on="CHAVE.I")
```

```python
#Corrigindo valores/missings para a proxima verificação
dados3.fillna(0,inplace=True)
```

```python
#Atribuindo valor médio do seguro desemprego para as pessoas selecionadas pela amostra
dados3.loc[(dados3["TEMP"]==1) & (dados3["SEGDEMP"]==0), "SEGDEMP"] = (1267.11*4.78)/12
```

```python
#Valor total do beneficio para cada individuo
dados3["BEN"]=dados3["SEGDEMP"]+dados3["SALFAM"]
```
